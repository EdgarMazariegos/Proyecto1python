```python
#Comenzando el modelo. Estamos utilizando Calidad

#calculando la predicción y^
b1=0
b0=200000
n=len(calidad_train)
lrate=0.05
epoch=np.linspace(1,5000,num=5000)


unos=np.ones_like(calidad_train).reshape(-1,1)
xmodelo=np.array(calidad_train).reshape(-1,1)
ymodelo=np.array(precioventa_train).reshape(-1,1)
beta_uno=np.array(np.ones_like(calidad_train)*b1).reshape(-1,1)
beta_cero=np.array(np.ones_like(calidad_train)*b0).reshape(-1,1)


x_unos=np.hstack([xmodelo,unos])
betas=np.hstack([beta_uno,beta_cero]).transpose()
prediccion=np.array(np.diag( x_unos.dot(betas))).reshape(-1,1)
sumatoriay=np.array((ymodelo-prediccion)**2)

err=((1/(2*n)))*(np.sum(sumatoriay))

sumgraderror1=np.array((prediccion-ymodelo)*xmodelo)
sumgraderror0=np.array(prediccion-ymodelo)

#gradientes de error
gerrorb1= (1/n)*(np.sum(sumgraderror1))
gerrorb0= (1/n)*(np.sum(sumgraderror0))


b1nuevo=b1-lrate*gerrorb1
b2nuevo=b0-lrate*gerrorb0
```
