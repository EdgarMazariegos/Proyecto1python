```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dd = np.load("proyecto_training_data.npy")

datos= pd.DataFrame(dd, columns= ["PrecioVenta", "Calidad", "Dimension", "Nhabitaciones", "Aconstruc", "piescalle"])
```

```python

#Dividiendo los datos de entrenamiento y prueba
y= datos.iloc[: , :1]
x= datos.iloc[: , 1:]


x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)


#dividiendo cada variable de train
precioventa_train=y_train.iloc[:,0]
calidad_train=x_train.iloc[ : ,0]
dimension_train=x_train.iloc[ : ,1]
Nhabitaciones_train=x_train.iloc[ : ,2]
Aconstruc_train=x_train.iloc[ : ,3]
piescalle_train=x_train.iloc[ : ,4]
```

```python
#Explorando valores de venta (y)

vmedia=np.mean(precioventa_train)
print(vmedia)
vmaximo=np.amax(precioventa_train)
print(vmaximo)
vminimo=np.amin(precioventa_train)
print(vminimo)
vrango=np.ptp(precioventa_train)
print(vrango)
vstd=np.std(precioventa_train)
print(vstd)

sns.displot(precioventa_train)
plt.show() #si tengo tiempo, arreglo los ejes de los histogramas de cada variable
```

```python
#Explorando valores de calidad 

cmedia=np.mean(calidad_train)
print(cmedia)
cmaximo=np.amax(calidad_train)
print(cmaximo)
cminimo=np.amin(calidad_train)
print(cminimo)
crango=np.ptp(calidad_train)
print(crango)
cstd=np.std(calidad_train)
print(cstd)

corrcalidad=np.array(np.corrcoef(calidad_train,precioventa_train))

#Histograma de calidad
sns.displot(calidad_train)
plt.show()


#Calculando el coeficiente de correlación entre calidad y precio de venta
plt.scatter(calidad_train, precioventa_train)
plt.title(f"r= {corrcalidad[0,1]}")
plt.xlabel("Valor de calidad")
plt.ylabel("Precio de venta")
plt.show()

```

```python
#Explorando valores de Dimension 

dmedia=np.mean(dimension_train)
print(dmedia)
dmaximo=np.amax(dimension_train)
print(dmaximo)
dminimo=np.amin(dimension_train)
print(dminimo)
drango=np.ptp(dimension_train)
print(drango)
dstd=np.std(dimension_train)
print(dstd)


corrdimension=np.array(np.corrcoef(dimension_train,precioventa_train))

#Histograma de dimension
sns.displot(dimension_train)
plt.show()


#Calculando el coeficiente de correlación entre dimension y precio de venta
plt.scatter(dimension_train, precioventa_train)
plt.title(f"r= {corrdimension[0,1]}")
plt.xlabel("Dimensiones")
plt.ylabel("Precio venta")
plt.show()

```

```python
#Explorando valores de Numero de habitaciones 

nmedia=np.mean(Nhabitaciones_train)
print(nmedia)
nmaximo=np.amax(Nhabitaciones_train)
print(nmaximo)
nminimo=np.amin(Nhabitaciones_train)
print(nminimo)
nrango=np.ptp(Nhabitaciones_train)
print(nrango)
nstd=np.std(Nhabitaciones_train)
print(nstd)

#Histograma de numero habitaciones
sns.displot(Nhabitaciones_train)
plt.show()

corrhabitaciones=np.array(np.corrcoef(Nhabitaciones_train,precioventa_train))


#Calculando el coeficiente de correlación entre numero habitaciones y precio de venta
plt.scatter(Nhabitaciones_train, precioventa_train)
plt.title(f"r= {corrhabitaciones[0,1]}")
plt.xlabel("N habitaciones")
plt.ylabel("Precio venta")
plt.show()

```

```python
#Explorando valores de Año de construcción

amedia=np.mean(Aconstruc_train)
print(amedia)
amaximo=np.amax(Aconstruc_train)
print(amaximo)
aminimo=np.amin(Aconstruc_train)
print(aminimo)
arango=np.ptp(Aconstruc_train)
print(arango)
astd=np.std(Aconstruc_train)
print(astd)

#Histograma Año de construcción
sns.displot(Aconstruc_train)
plt.show()


corraconstruct=np.array(np.corrcoef(Aconstruc_train,precioventa_train)) 
print()

#Calculando el coeficiente de correlación entre año de construcción y precio de venta
plt.scatter(Aconstruc_train, precioventa_train)
plt.title(f"r= {corraconstruct[0,1]}")
plt.xlabel("Año construcción")
plt.ylabel("Precio venta")
plt.show()

```

```python
#Explorando valores de pies de calle conectados a la propiedad

pmedia=np.mean(piescalle_train)
print(pmedia)
pmaximo=np.amax(piescalle_train)
print(pmaximo)
pminimo=np.amin(piescalle_train)
print(pminimo)
prango=np.ptp(piescalle_train)
print(prango)
pstd=np.std(piescalle_train)
print(pstd)

#Histograma pies lineares de calle conectados a casa
sns.displot(piescalle_train)
plt.show()


corrpiescalle=np.array(np.corrcoef(piescalle_train,precioventa_train)) 


#Calculando el coeficiente de correlación entre pies lineares de calle y precio de venta
plt.scatter(piescalle_train, precioventa_train)
plt.title(f"r= {corrpiescalle[0,1]}")
plt.xlabel("Pies de calle conectados a casa")
plt.ylabel("Precio venta")
plt.show()



```
