```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dd = np.load("proyecto_training_data.npy")

datos= pd.DataFrame(dd, columns= ["PrecioVenta", "Calidad", "Dimension", "Nhabitaciones", "Aconstruc", "piescalle"])
```

```python

#Dividiendo los datos de entrenamiento y prueba
y= datos.iloc[: , :1]
x= datos.iloc[: , 1:]

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)


#dividiendo cada variable de train
precioventa_train=y_train.iloc[:,0]
calidad_train=x_train.iloc[ : ,0]
dimension_train=x_train.iloc[ : ,1]
Nhabitaciones_train=x_train.iloc[ : ,2]
Aconstruc_train=x_train.iloc[ : ,3]
piescalle_train=x_train.iloc[ : ,4]
```

```python
#Explorando valores de venta (y)

vmedia=np.mean(precioventa_train)
print(vmedia)
vmaximo=np.amax(precioventa_train)
print(vmaximo)
vminimo=np.amin(precioventa_train)
print(vminimo)
vrango=np.ptp(precioventa_train)
print(vrango)
vstd=np.std(precioventa_train)
print(vstd)

sns.displot(precioventa_train)
plt.show() #si tengo tiempo, arreglo los ejes de los histogramas de cada variable
```

```python
#Explorando valores de calidad 

cmedia=np.mean(calidad_train)
print(cmedia)
cmaximo=np.amax(calidad_train)
print(cmaximo)
cminimo=np.amin(calidad_train)
print(cminimo)
crango=np.ptp(calidad_train)
print(crango)
cstd=np.std(calidad_train)
print(cstd)

corrcalidad=np.array(np.corrcoef(calidad_train,precioventa_train))

#Histograma de calidad
sns.displot(calidad_train)
plt.show()


#Calculando el coeficiente de correlación entre calidad y precio de venta
plt.scatter(calidad_train, precioventa_train)
plt.title(f"r= {corrcalidad[0,1]}")
plt.xlabel("Precio de venta")
plt.ylabel("Valor de calidad")
plt.show()

```
